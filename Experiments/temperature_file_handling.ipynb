```python
import re
```

```python
def parse_for_temperature(path):
    """Given a temperature file, parses it for current temp.
    
    Args:
        path (str): Path to temperature file.
    """
    with open(path, 'r') as f:
        for line in f:
            match = re.search(r't=(\d+)', line)
            if match:
                temp = match.groups()[0]
    return temp
```

```python
parse_for_temperature(temperature_file_path)
```

```python
import subprocess
import os
```

```python
def get_temperature_file(local_dir, pi_ip='192.168.43.127'):
    """Copies temperature file from the raspberry pi to local system.
    
    Args:
        local_dir (str): Path to directory to copy `w1_slave` file too.
        pi_ip (str, optional): IP address for the pi
    """
    file_path = local_dir + '/w1_slave'
    if os.path.isfile(file_path):
        os.remove(file_path)
    command = ["scp", 
              "pi@{}:/sys/bus/w1/devices/28-00000597c7ff/w1_slave".format(pi_ip),
              local_dir]
    subprocess.call(command)
    return local_dir + '/w1_slave'
```
